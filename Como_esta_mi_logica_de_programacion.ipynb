##üß© Ejercicio 1: Sistema de riego inteligente
###`Una finca agr√≠cola desea automatizar el proceso de riego utilizando datos de humedad del suelo y pron√≥stico de lluvia. Se pide desarrollar un programa que analice los niveles de humedad diaria de una semana (expresados en porcentaje) y determine si es necesario activar el sistema de riego cada d√≠a. Para ello, el sistema debe considerar un umbral de humedad √≥ptima y un margen de seguridad definidos como constantes, y evitar regar si se pronostica lluvia. El programa deber√° recorrer los siete d√≠as de la semana con un bucle, aplicar condicionales para decidir cu√°ndo regar y mostrar un resumen final con el n√∫mero total de riegos realizados y el ahorro de agua estimado.`

```python
umbral=float(input("Digite el umbral de humedad"))
margen=float(input("Digite el margen de seguridad de lluevia"))
limite_agua=float(input("Digite la cantidad de agua por todo la semana"))
agua_gastada=float(input("Digite la cantidad de agua que se gasta por regar"))
dias=["lunes","martes","miercoles","jueves","viernes","sabado","domingo"]
resultado=[]
gasto=0
ahorro=0
for dia in dias:
    humedad_suelo=float(input(f"Digite la humedad diaria en decimal del dia {dia}:"))
    pronostico_lluvia=float(input("Digite el pron√≥stico de lluvia"))
    if humedad_suelo<umbral and pronostico_lluvia<margen:
        limite_agua-=agua_gastada
        gasto+=agua_gastada
        resultado.append([f"El dia {dia} tuvo un humedad de {humedad_suelo*100}% y un pronostico de lluvia de {pronostico_lluvia*100}% por lo tanto se recomienda regar agua gastada {agua_gastada} en el riego, cantidad que a√∫n queda para la semana:{limite_agua}"])
    else:
      ahorro_agua=+limite_agua
      ahorro+=agua_gastada
      resultado.append([f"El dia {dia} tuvo un humedad de {humedad_suelo*100}% y un pronostico de lluvia de {pronostico_lluvia*100}% por lo tanto no se recomienda regar el dia {dia}, ahorro de {agua_gastada}, cantidad que queda{ahorro_agua}"])
for salida in resultado:
    print(salida[0])
print(f"RESULTADO:")
print(f"el total de riego realizado fue de {gasto}")
print(f"el ahorro de agua fue de {ahorro}")





```

El programa debe usar funciones para los c√°lculos, bucles para procesar varios lotes y condicionales para establecer la categor√≠a final, mostrando un reporte con los resultados.`

##üåæ Ejercicio 3: Planificaci√≥n de rotaci√≥n de cultivos
###`Un ingeniero agr√≥nomo busca dise√±ar una herramienta que ayude a planificar la rotaci√≥n de cultivos durante el a√±o y calcular el consumo total de nitr√≥geno (N) en el suelo. El programa deber√° recibir una lista con los cultivos sembrados cada mes (por ejemplo, ma√≠z, soya, trigo o descanso) y usar constantes que indiquen el consumo mensual de nitr√≥geno por cultivo. Mediante funciones, el sistema debe sumar el consumo total de N y verificar si existen m√°s de tres meses consecutivos del mismo cultivo, recomendando alternar con un cultivo diferente o un per√≠odo de descanso.`

```python
CONSUMO_NITROGENO = {
    "maiz": 30,
    "soya": 20,
    "trigo": 25,
    "descanso": 0
}

def calcular_consumo_total(cultivos):
    total = 0
    for cultivo in cultivos:
        total += CONSUMO_NITROGENO.get(cultivo.lower(), 0)
    return total

def verificar_cultivo(cultivos):
    recomendaciones = []
    cultivo_actual = None
    contador = 0

    for i, cultivo in enumerate(cultivos):
        if cultivo == cultivo_actual:
            contador += 1
        else:
            cultivo_actual = cultivo
            contador = 1

        if contador >= 3:
            recomendaciones.append(
                f"Mes {i+1}: m√°s de 3 meses consecutivos de {cultivo}. Se recomienda alternar o descansar."
            )
    return recomendaciones

meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
         "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

cultivos = []
for mes in meses:
    cultivo = input(f"Ingrese el cultivo para {mes} (maiz, soya, trigo o descanso): ").strip().lower()
    cultivos.append(cultivo)

consumo_total = calcular_consumo_total(cultivos)
alertas = verificar_cultivo(cultivos)

print("\n Planificaci√≥n anual de cultivos:")
for i in range(12):
    print(f"{meses[i]}: {cultivos[i].capitalize()} ({CONSUMO_NITROGENO.get(cultivos[i], 0)} kg de N)")

print(f"\n Consumo total de nitr√≥geno en el a√±o: {consumo_total} kg/ha")

if alertas:
    print("\n Recomendaciones de rotaci√≥n:")
    for alerta in alertas:
        print(alerta)
else:
    print("\n No se detectaron cultivos prolongados.")

```

##üíª Ejercicio 4: Control de inventario de un software agr√≠cola
###`Una empresa que desarrolla software para el agro necesita un m√≥dulo que controle el inventario de insumos (fertilizantes, semillas, pesticidas). El programa debe permitir registrar la cantidad inicial de cada producto y simular la salida de unidades seg√∫n las ventas semanales. Con el uso de bucles y condicionales, el sistema debe actualizar los valores y emitir una alerta cuando alg√∫n producto quede por debajo de un nivel m√≠nimo definido como constante. Adem√°s, se debe incluir una funci√≥n que calcule el valor total del inventario restante y muestre un reporte con el stock actualizado.`



```python
NIVEL_MINIMO = 10

inventario = {}
n = int(input("¬øCu√°ntos productos tiene el inventario? "))

for _ in range(n):
    nombre = input("\nNombre del producto: ").lower()
    cantidad = int(input(f"Cantidad disponible de {nombre}: "))
    precio = float(input(f"Precio por unidad de {nombre}: "))
    inventario[nombre] = {"cantidad": cantidad, "precio": precio}

ventas_semanales = {}
print("\n--- Ingrese las ventas semanales ---")

for producto in inventario:
    ventas_producto = []
    semanas = int(input(f"\n¬øCu√°ntas semanas de ventas para '{producto}'?: "))
    for i in range(semanas):
        cantidad = int(input(f"  Ventas de la semana {i+1} para '{producto}': "))
        ventas_producto.append(cantidad)
    ventas_semanales[producto] = ventas_producto

def actualizar_inventario(inventario, ventas):
    for producto, salidas in ventas.items():
        for cantidad_vendida in salidas:
            inventario[producto]["cantidad"] -= cantidad_vendida
            if inventario[producto]["cantidad"] < NIVEL_MINIMO:
                print(f"Alerta: El stock de '{producto}' est√° por debajo del nivel m√≠nimo "
                      f"({inventario[producto]['cantidad']} unidades).")

def calcular_valor_total(inventario):
    total = 0
    for producto, datos in inventario.items():
        total += datos["cantidad"] * datos["precio"]
    return total

def mostrar_reporte(inventario):
    print("\n Reporte de Inventario Actualizado:")
    for producto, datos in inventario.items():
        print(f"- {producto.capitalize()}: {datos['cantidad']} unidades restantes")
    print(f"\n Valor total del inventario: ${calcular_valor_total(inventario)}")

actualizar_inventario(inventario, ventas_semanales)
mostrar_reporte(inventario)
1
```

##‚öôÔ∏è Ejercicio 5: Evaluaci√≥n de desempe√±o de maquinaria
###`Una cooperativa agr√≠cola quiere medir el rendimiento de sus tractores en campo. Crea un programa que permita ingresar los datos de varios tractores: horas trabajadas, litros de combustible consumidos y superficie cubierta. El programa debe calcular el rendimiento por hora y el consumo de combustible por hect√°rea. Mediante condicionales, debe clasificar el desempe√±o como ‚ÄúExcelente‚Äù, ‚ÄúAceptable‚Äù o ‚ÄúDeficiente‚Äù, seg√∫n los valores obtenidos. Usa funciones para los c√°lculos, bucles para procesar varios tractores y constantes para los l√≠mites de clasificaci√≥n. Finalmente, muestra un informe con los resultados de cada m√°quina.`

```python
RENDIMIENTO_EXCELENTE = 1.5
RENDIMIENTO_ACEPTABLE = 1.0
CONSUMO_EFICIENTE = 10
CONSUMO_ACEPTABLE = 15

def calcular_rendimiento_por_hora(superficie, horas):
    if horas > 0:
        return superficie / horas
    elif horas == 0:
        return 0
    else:
        print("Error: Las horas no pueden ser negativas.")
        return 0

def calcular_consumo_por_hectarea(combustible, superficie):
    if superficie > 0:
        return combustible / superficie
    elif superficie == 0:
        return 0
    else:
        print("Error: La superficie no puede ser negativa.")
        return 0

def clasificar_desempe√±o(rendimiento, consumo):
    if rendimiento >= RENDIMIENTO_EXCELENTE and consumo <= CONSUMO_EFICIENTE:
        return "Excelente"
    elif rendimiento >= RENDIMIENTO_ACEPTABLE and consumo <= CONSUMO_ACEPTABLE:
        return "Aceptable"
    else:
        return "Deficiente"

def evaluar_tractores():
    cantidad = int(input("¬øCu√°ntos tractores deseas evaluar? "))
    resultados = []

    for i in range(cantidad):
        print(f"\nTractor #{i + 1}")
        horas = float(input("Horas trabajadas: "))
        combustible = float(input("Litros de combustible consumidos: "))
        superficie = float(input("Superficie cubierta (hect√°reas): "))

        rendimiento = calcular_rendimiento_por_hora(superficie, horas)
        consumo = calcular_consumo_por_hectarea(combustible, superficie)
        desempe√±o = clasificar_desempe√±o(rendimiento, consumo)

        resultados.append({
            "Tractor": i + 1,
            "Rendimiento (ha/h)": round(rendimiento, 2),
            "Consumo (L/ha)": round(consumo, 2),
            "Desempe√±o": desempe√±o
        })

    print("\n Informe de Desempe√±o:")
    for r in resultados:
        print(f"Tractor {r['Tractor']}: Rendimiento = {r['Rendimiento (ha/h)']} ha/h, "
              f"Consumo = {r['Consumo (L/ha)']} L/ha, Desempe√±o: {r['Desempe√±o']}")

evaluar_tractores()
```

##üöó Ejercicio 3: Optimizaci√≥n de rutas de reparto:
###`Una empresa de mensajer√≠a quiere optimizar sus rutas diarias.`
###`Crea un programa que reciba las distancias (en km) y los tiempos estimados (en minutos) de varios trayectos.`
###`El sistema debe:`

###-`Calcular la velocidad promedio de cada trayecto.`

###-`Determinar cu√°l es la ruta m√°s eficiente (mayor velocidad promedio con menor distancia).`

###-`Calcular el tiempo total del recorrido completo.`
###`-Usa funciones para los c√°lculos, bucles para recorrer las rutas y condicionales para la selecci√≥n de la mejor opci√≥n, mostrando un informe final con las m√©tricas.`

```python
def calcular_velocidad_promedio(distancia, tiempo_min):
    tiempo_horas = tiempo_min / 60
    velocidad = distancia / tiempo_horas
    return velocidad

def ruta_mas_eficiente(distancias, tiempos):
    velocidades = []
    for i in range(len(distancias)):
        velocidad = calcular_velocidad_promedio(distancias[i], tiempos[i])
        velocidades.append(velocidad)

    mejor_indice = 0
    for i in range(1, len(distancias)):
        if velocidades[i] > velocidades[mejor_indice]:
            mejor_indice = i
        elif velocidades[i] == velocidades[mejor_indice] and distancias[i] < distancias[mejor_indice]:
            mejor_indice = i

    return mejor_indice, velocidades

def tiempo_total(tiempos):
    return sum(tiempos)

def generar_informe(distancias, tiempos, velocidades, mejor_indice):
    print("\nINFORME DE OPTIMIZACI√ìN DE RUTAS:")
    for i in range(len(distancias)):
        print(f"Ruta {i + 1}: {distancias[i]} km, {tiempos[i]} min -> Velocidad promedio: {velocidades[i]:.2f} km/h")

    print("\nRuta m√°s eficiente:")
    print(f"Ruta {mejor_indice + 1} con {distancias[mejor_indice]} km, {tiempos[mejor_indice]} min, "
          f"velocidad promedio de {velocidades[mejor_indice]:.2f} km/h")

    print(f"\nTiempo total de todas las rutas: {tiempo_total(tiempos)} minutos")

print("Rutas de Reparto:")
num_rutas = int(input("\nIngrese el n√∫mero de rutas a evaluar: "))

distancias = []
tiempos = []

for i in range(num_rutas):
    print(f"\nRuta {i + 1}:")
    distancia = float(input("Distancia (km): "))
    tiempo = float(input("Tiempo estimado (minutos): "))
    distancias.append(distancia)
    tiempos.append(tiempo)

mejor_indice, velocidades = ruta_mas_eficiente(distancias, tiempos)
generar_informe(distancias, tiempos, velocidades, mejor_indice)
```
