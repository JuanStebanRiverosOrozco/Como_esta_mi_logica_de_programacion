##üß© Ejercicio 1: Sistema de riego inteligente
###`Una finca agr√≠cola desea automatizar el proceso de riego utilizando datos de humedad del suelo y pron√≥stico de lluvia. Se pide desarrollar un programa que analice los niveles de humedad diaria de una semana (expresados en porcentaje) y determine si es necesario activar el sistema de riego cada d√≠a. Para ello, el sistema debe considerar un umbral de humedad √≥ptima y un margen de seguridad definidos como constantes, y evitar regar si se pronostica lluvia. El programa deber√° recorrer los siete d√≠as de la semana con un bucle, aplicar condicionales para decidir cu√°ndo regar y mostrar un resumen final con el n√∫mero total de riegos realizados y el ahorro de agua estimado.`

```python
umbral=float(input("Digite el umbral de humedad"))
margen=float(input("Digite el margen de seguridad de lluevia"))
limite_agua=float(input("Digite la cantidad de agua por todo la semana"))
agua_gastada=float(input("Digite la cantidad de agua que se gasta por regar"))
dias=["lunes","martes","miercoles","jueves","viernes","sabado","domingo"]
resultado=[]
gasto=0
ahorro=0
for dia in dias:
    humedad_suelo=float(input(f"Digite la humedad diaria en decimal del dia {dia}:"))
    pronostico_lluvia=float(input("Digite el pron√≥stico de lluvia"))
    if humedad_suelo<umbral and pronostico_lluvia<margen:
        limite_agua-=agua_gastada
        gasto+=agua_gastada
        resultado.append([f"El dia {dia} tuvo un humedad de {humedad_suelo*100}% y un pronostico de lluvia de {pronostico_lluvia*100}% por lo tanto se recomienda regar agua gastada {agua_gastada} en el riego, cantidad que a√∫n queda para la semana:{limite_agua}"])
    else:
      ahorro_agua=+limite_agua
      ahorro+=agua_gastada
      resultado.append([f"El dia {dia} tuvo un humedad de {humedad_suelo*100}% y un pronostico de lluvia de {pronostico_lluvia*100}% por lo tanto no se recomienda regar el dia {dia}, ahorro de {agua_gastada}, cantidad que queda{ahorro_agua}"])
for salida in resultado:
    print(salida[0])
print(f"RESULTADO:")
print(f"el total de riego realizado fue de {gasto}")
print(f"el ahorro de agua fue de {ahorro}")





```

##üçì Ejercicio 2: Clasificaci√≥n de calidad de cosecha
###`Una cooperativa frut√≠cola necesita clasificar sus lotes de frutas seg√∫n la calidad, en base al peso promedio y al porcentaje de piezas defectuosas. Se requiere crear un programa que, para varios lotes ingresados por el usuario, calcule el peso promedio de una muestra y determine el porcentaje de defectos. En funci√≥n de estos valores, el sistema debe asignar una categor√≠a: ‚ÄúA‚Äù para productos de alta calidad, ‚ÄúB‚Äù para calidad media y ‚ÄúRechazo‚Äù para los que no cumplen los est√°ndares. El programa debe usar funciones para los c√°lculos, bucles para procesar varios lotes y condicionales para establecer la categor√≠a final, mostrando un reporte con los resultados.`



```python
numero_lotes = int(input("Ingrese el n√∫mero de lotes totales: "))
def calcular_peso_promedio(peso_fruta,peso_total=0):
  peso_total = peso_total + peso_fruta
  peso_promedio = peso_total / numero_lotes
  return peso_promedio

def calcular_porcentaje_defectos(porcentaje_defectos):
  porcentaje_defectos = porcentaje_defectos / numero_lotes
  return porcentaje_defectos

def determinar_categoria(peso_promedio, porcentaje_defectos):
  if peso_promedio > 100 and porcentaje_defectos < 5:
    categoria = "A"
  elif 50 < peso_promedio < 100 and porcentaje_defectos < 10:
    categoria = "B"
  else:
    categoria = "Rechazo"
  return categoria

for lote in range(numero_lotes):
  peso_fruta = float(input("Ingrese el peso de la fruta: "))
  porcentaje_defectos = float(input("Ingrese el porcentaje de defectos: "))

peso_promedio = calcular_peso_promedio(peso_fruta)
porcentaje_defectos = calcular_porcentaje_defectos(porcentaje_defectos)
categoria = determinar_categoria(peso_promedio, porcentaje_defectos)

print("RESULTADO:")
print("Peso promedio:", peso_promedio)
print("Porcentaje de defectos:", porcentaje_defectos)
print("Categor√≠a:", categoria)
```
